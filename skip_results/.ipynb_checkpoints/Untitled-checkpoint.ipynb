```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
skip = pd.read_csv('2shot_skip_blackswan.csv')
no_skip = pd.read_csv('2shot_blackswan.csv')
```

```python
plt.figure(figsize=(16, 6))

plt.subplot(121)
plt.plot(skip['Wall time'] - skip['Wall time'][0], skip['Value'], label='Skip')
plt.plot(no_skip['Wall time'] - no_skip['Wall time'][0], no_skip['Value'], label='No Skip')
plt.legend()
plt.xlabel('time', fontsize=16)

plt.subplot(122)
plt.plot(skip['Value'], label='Skip')
plt.plot(no_skip['Value'], label='No Skip')
plt.legend()
plt.xlabel('epoch', fontsize=16)

plt.show()
```
